```python
import pandas as pd

df = pd.read_csv('train.csv').drop(['id'], axis = 1)
test = pd.read_csv('test.csv').drop(['id'], axis = 1)
sample = pd.read_csv('sample_submission.csv')
df
```

```python
df['Exter Qual'].unique()
df['Bsmt Qual'].unique()
df['Kitchen Qual'].unique()
```

```python
dict1 = {'Ex' : 1, 'Gd': 2, 'TA':3, 'Fa':4}
dict2 = {'Ex' : 1, 'Gd': 2, 'TA':3, 'Fa':4, 'Po':5}
dict3 = {'Ex' : 1, 'Gd': 2, 'TA':3, 'Fa':4, 'Po':5}
df['Q'] = 0
for i in range(len(df)):
    df['Exter Qual'][i] = dict1[df['Exter Qual'][i]]
    df['Bsmt Qual'][i] = dict2[df['Bsmt Qual'][i]]
    df['Kitchen Qual'][i] = dict3[df['Kitchen Qual'][i]]
    df['Q'][i] = (df['Exter Qual'][i] + df['Bsmt Qual'][i] + df['Kitchen Qual'][i])//3
df
```

```python
test['Q'] = 0
for i in range(len(test)):
    test['Exter Qual'][i] = dict1[test['Exter Qual'][i]]
    test['Bsmt Qual'][i] = dict2[test['Bsmt Qual'][i]]
    test['Kitchen Qual'][i] = dict3[test['Kitchen Qual'][i]]
    test['Q'][i] = (test['Exter Qual'][i] + test['Bsmt Qual'][i] + test['Kitchen Qual'][i])//3
test
```

# 모델설정

```python
from pycaret.regression import *

setup_ = setup(data = df, use_gpu = True, target = 'target', n_jobs = -1, train_size = 0.9, feature_interaction = True, feature_ratio = True)
```

```python
best5 = compare_models(n_select = 5, sort = 'MAE', fold = 10)
best5
```

# 학습

```python
cat = create_model('catboost')
```

```python
cat.get_all_params()
```

# 모델튜닝

```python
# params = {"depth": [3,4,5,6,7,8,9],
#           "learning_rate": [0.04, 0.06],
#           "iterations": [1000],
#           }
```

```python
# tuned_cat = tune_model(cat, custom_grid = params, optimize='MAE',)
tuned_cat = tune_model(cat, optimize='MAE',)
```

```python
tuned_cat.get_all_params()
```

```python
bagging = ensemble_model(tuned_cat, method = 'Bagging')
boosting = ensemble_model(tuned_cat, method = 'Boosting')
```

```python
predict_model(bagging)
```

```python
predict_model(boosting)
```

```python
blender = blend_models(estimator_list = [bagging, boosting, tuned_cat])
```

```python
predict_model(blender)
```

```python
stacknet = stack_models(estimator_list = [tuned_cat, boosting, bagging, blender],)
```

```python
predict_model(stacknet)
```

# Finalize

```python
final_model = finalize_model(blender)
```

```python
prediction = predict_model(final_model, data=test)
```
